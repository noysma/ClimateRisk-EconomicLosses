```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline 
```

```python
df = pd.read_csv("climate-risk-index.csv")
```

### Drop Null columns and values

```python
df = df.drop(columns=["the_geom", "the_geom_webmercator", "cartodb_id"])
df = df.dropna()
```

# Comparing a nation's climate risk ranking against nations

```python
plt.figure(figsize=(15,25))
plt.title("CRI per Country\n", fontsize = 30, fontweight="bold")
sns.barplot(data=df, x="cri_rank", y="rw_country_name",
                        order=df.sort_values('cri_rank').rw_country_name,
                        palette=sns.color_palette('coolwarm', n_colors=len(df)))
plt.xlabel('\nClimate Risk Index', fontsize=20, style="italic")
plt.ylabel('Country Name\n', fontsize=20, style="italic")

plt.show()
```

## 15 Less Affected Country

```python
plt.figure(figsize=(15,8))
plt.title("CRI per Country\n", fontsize = 20, fontweight="bold")
ax = sns.barplot(data=df, x="cri_rank", y="rw_country_name",
                        order=df.sort_values('cri_rank').rw_country_name[-15:],
                        palette=sns.color_palette('coolwarm', n_colors=15))
ax.bar_label(ax.containers[0])
plt.xlabel('\nClimate Risk Index Rank', fontsize=15, style="italic")
plt.ylabel('Country Name\n', fontsize=15, style="italic")

plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.title("CRI per Country\n", fontsize = 20, fontweight="bold")
ax = sns.barplot(data=df, x="cri_score", y="rw_country_name",
                        order=df.sort_values('cri_rank').rw_country_name[-15:],
                        palette=sns.color_palette('coolwarm', n_colors=15))
ax.bar_label(ax.containers[0])
plt.xlabel('\nClimate Risk Index Score', fontsize=15, style="italic")
plt.ylabel('Country Name\n', fontsize=15, style="italic")

plt.show()
```
